```python
import pandas as pd
import altair as alt
import re
```

```python
sge = '../Data/20250821_BARD1scores_newmodel_wGMM_0.999_FILTERED.xlsx'
```

```python
def read_data(file):
    df = pd.read_excel(file)

    df['flip'] = df['functional_consequence'] + '->' + df['gmm_consequence']

    df = df.loc[~(df['flip'].isin(['functionally_normal->functionally_normal', 'functionally_abnormal->functionally_abnormal', 'indeterminate->indeterminate']))]

    df.to_excel('/Users/ivan/Desktop/test_excel_outputs/20250822_BARD1_newmodel_ZScorevsGMM.xlsx', index = False)
    coding_vars = df.loc[df['consequence'].isin(['synonymous_variant', 'stop_gained', 'missense_variant'])]
    coding_vars = coding_vars.copy()
    coding_vars['aa_pos'] = coding_vars['amino_acid_change'].transform(lambda x: int(x[1:-1]))

    
    return coding_vars
```

```python
def visualize(df):
    chart = alt.Chart(df).mark_tick().encode(
        x = 'aa_pos',
        y = 'flip',
        color = 'consequence',
        tooltip = ['amino_acid_change']
    ).properties(
        width = 600, 
        height = 400
    ).interactive()

    chart.display()
```

```python
def main():
    coding_df = read_data(sge)
    visualize(coding_df)
```

```python
main()
```
