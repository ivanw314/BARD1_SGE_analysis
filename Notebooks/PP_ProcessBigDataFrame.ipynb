This notebook processes the large pillar project dataframe and extracts data relevant to figure 2. There are three Excel file outputs: 1) a complete file with all data from SGE and VAMP-seq, 2) only SGE data, and 3) only VAMP-seq data

```python
import pandas as pd
```

```python
save = False
bigppj_df = '' #Path to pillar project dataframe

sge_genes = ['BARD1', 'PALB2', 'BRCA2', 'RAD51D', 'XRCC2', 'CTCF', 'SFPQ']
vampseq_genes = ['G6PD', 'TSC2', 'F9']
```

```python
df = pd.read_csv(bigppj_df) #Reads the dataframe


sge_subset = df.loc[(df['Gene'].isin(sge_genes)) & (df['Dataset'].str.contains('_unpublished'))] #Gets all SGE data 
vampseq_subset = df.loc[df['Gene'].isin(vampseq_genes)] #Pulls as VAMP-seq data


#For VAMP-seq F9 data, the heavy chain antibody data is used
mask = df['Gene'] == 'F9'
vampseq_subset = vampseq_subset[(vampseq_subset['Gene'] != 'F9') | (vampseq_subset['Dataset'].str.contains('heavy'))]


if save:
    print('Saving SGE...')
    sge_subset.to_excel('../Data/filtered_ppj_data/20251202_SGEsubset.xlsx', index = False)
    
    print('Saving VAMP-seq...')
    vampseq_subset.to_excel('../Data/filtered_ppj_data/20251204_VAMPseqsubset_wDups.xlsx', index = False)
    
    
    print('Saving CAVA subset...')
    cava_subset = pd.concat([sge_subset, vampseq_subset])
    cava_subset.to_excel('../Data/filtered_ppj_data/20251202_CAVAseqsubset.xlsx', index = False)

    print('Done!')

else:
    print(f'No files saved')
```
