```python
import pandas as pd
import altair as alt
from natsort import natsorted
```

```python
file = '/Users/ivan/Downloads/BARD1.editrates.sorted.tsv'
```

```python
def read_process(file):
    df = pd.read_csv(file, sep = '\t')

    df['target'] = df['target_rep'].transform(lambda x: x.split('X')[1].split('_')[0])
    df['rep'] = df['target_rep'].transform(lambda x: x.split('X')[1].split('_')[1])

    df.loc[(df['rep'] == 'R1R4') | (df['rep'] == 'R1R2R3'), 'rep'] = 'Rep. 1'
    df.loc[(df['rep'] == 'R2R5') | (df['rep'] == 'R4R5R6'), 'rep'] = 'Rep. 2'
    df.loc[(df['rep'] == 'R3R6') | (df['rep'] == 'R7R8R9'), 'rep'] = 'Rep. 3'

    
    return df
```

```python
def bar_plot(df):

    sort_order = natsorted(df['target'].tolist())
    plot = alt.Chart(df).mark_bar().encode(
        x = alt.X('rep',
                  axis = alt.Axis(
                      title = '',
                      labels = False,
                      ticks = False
                  )
                 ),
        y = alt.Y('edit_rate',
                  title = 'Lib. Edit Rate',
                  scale = alt.Scale(domain = [0, 0.5])
                 ),
        column = alt.Column('target',
                            sort = sort_order,
                            header=alt.Header(title='SGE Target')
                           ),
        color = alt.Color('rep', 
                          legend = alt.Legend(
                              title = ''
                          )
                         )
    ).properties(
        width = 25,
        height = 200
    ).configure_facet(
        spacing = 5
    ).configure_axis(
        grid = False
    )

    plot.display()
    #plot.save('/Users/ivan/Desktop/BARD1_draft_figs/supp_figs/EditRate_barplot.png', dpi = 400)
```

```python
def main():
    df = read_process(file)
    bar_plot(df)
```

```python
main()
```
