```python
import pandas as pd
```

```python
bigppj_df = '/Users/ivan/Downloads/final_pillar_data_with_clinvar_gnomad_wREVEL_wAM_wspliceAI_wMutpred2_wtrainvar_expanded_111225.csv'

sge_genes = ['BARD1', 'PALB2', 'BRCA2', 'RAD51D', 'XRCC2', 'CTCF', 'SFPQ']
vampseq_genes = ['G6PD', 'TSC2', 'F9']
```

```python
df = pd.read_csv(bigppj_df)


sge_subset = df.loc[(df['Gene'].isin(sge_genes)) & (df['Dataset'].str.contains('_unpublished'))]
vampseq_subset = df.loc[df['Gene'].isin(vampseq_genes)]


mask = df['Gene'] == 'F9'
vampseq_subset = vampseq_subset[(vampseq_subset['Gene'] != 'F9') | (vampseq_subset['Dataset'].str.contains('heavy'))]

vampseq_subset = vampseq_subset.drop_duplicates(subset = ['auth_reported_score'])

#print('Saving SGE...')
#sge_subset.to_excel('../Data/filtered_ppj_data/20251202_SGEsubset.xlsx', index = False)

print('Saving VAMP-seq...')
vampseq_subset.to_excel('../Data/filtered_ppj_data/20251202_VAMPseqsubset.xlsx', index = False)

print('Saving CAVA subset...')
cava_subset = pd.concat([sge_subset, vampseq_subset])
cava_subset.to_excel('../Data/filtered_ppj_data/20251202_CAVAseqsubset.xlsx', index = False)
print('Done!')
```
