```python
def read_scores(file): #reads SGE scores
    
    df = pd.read_csv(file, sep = '\t')
    df = df.rename(columns = {'score': 'snv_score', 'simplified_consequence': 'Consequence'})
    df = df[['exon','Consequence','snv_score']]
    df = df.loc[~(df['Consequence'].isin(['stop_lost']))]

    return df
```

```python
def prep_data(df): #Renames categories to be nicer
    
    df = df.dropna(subset = ['simplified_consequence'])
    df = df.rename(columns = {'simplified_consequence': 'Consequence',
                             'auth_reported_score': 'snv_score',
                              'auth_reported_func_class': 'functional_consequence',
                             'ref_allele': 'ref'})

    df.loc[df['Consequence'].str.contains('missense'), 'Consequence'] = 'Missense'
    df.loc[df['Consequence'] == 'synonymous_variant', 'Consequence'] = 'Synonymous'
    df.loc[df['Consequence'] == 'intron_variant', 'Consequence'] = 'Intron'
    df.loc[df['Consequence'] == 'stop_gained', 'Consequence'] = 'Stop Gained'
    df.loc[df['Consequence'] == 'stop_lost', 'Consequence'] = 'Stop Lost'
    df.loc[df['Consequence'].str.contains('site'), 'Consequence'] = 'Canonical Splice'
    df.loc[df['Consequence'].str.contains('ing_var'), 'Consequence'] = 'Splice Region'
    df.loc[df['Consequence'].str.contains('UTR'), 'Consequence'] = 'UTR Variant'
    df.loc[df['Consequence'] == 'start_lost', 'Consequence'] = 'Start Lost'
    df.loc[df['ref'].str.len() > 1, 'Consequence'] = '3bp Deletion'
   
    return df
```

```python
def compute_density(df, category): #used to calculate the density - mostly GPT code
    values = df['snv_score'].values
    density = gaussian_kde(values)
    xs = np.linspace(values.min(), values.max(), 200)
    ys = density(xs)
    
    return pd.DataFrame({'SGE_score': xs, 'density': ys, 'Consequence': category})
```

```python
def make_ridgeline_plot(df, rawdf):
    """Create a ridgeline plot showing SGE score distributions by variant consequence."""
    
    # Combine density and raw data
    combined_df = pd.concat([df, rawdf], ignore_index=True)
    
    # Define color palette and sort order
    PALETTE = [
        '#006616',  # dark green - Synonymous
        '#81B4C7',  # dusty blue - Missense
        '#ffcd3a',  # yellow - Stop Gained
        '#6AA84F',  # med green - Intron
        '#93C47D',  # light green - UTR Variant
        '#888888',  # med gray - Stop Lost
        '#000000',  # black - Start Lost
        '#1170AA',  # darker blue - Canonical Splice
        '#CFCFCF',  # light gray - Splice Region
        '#FF9A00'   # orange - 3bp Deletion
    ]
    
    VARIANT_ORDER = [
        'Synonymous', 'Missense', 'Stop Gained', 'Intron', 'UTR Variant',
        'Stop Lost', 'Start Lost', 'Canonical Splice', 'Splice Region', '3bp Deletion'
    ]
    
    # Create density plot with black outline
    base = alt.Chart(combined_df).mark_area(
        interpolate='monotone',
        opacity=0.7,
        line={'color': 'black', 'strokeWidth': 1.5}
    ).encode(
        x=alt.X('SGE_score:Q', 
                axis=alt.Axis(
                    title='SGE Score', 
                    titleFontSize=20, 
                    labelFontSize=16,
                    orient='bottom'
                )),
        y=alt.Y('density:Q', 
                axis=None),
        color=alt.Color('Consequence:N', 
                        legend=None,
                        scale=alt.Scale(range=PALETTE, domain=VARIANT_ORDER)),
        order=alt.Order('density:Q')
    ).properties(
        height=40,
        width=1000
    )
    
    # Facet by consequence
    plot = base.facet(
        row=alt.Row('Consequence:N', 
                    title=None,
                    sort=VARIANT_ORDER,
                    header=alt.Header(
                        labelAngle=0, 
                        labelFontSize=16,
                        labelAlign='right',
                        labelBaseline='middle',
                        titleOrient='left'
                    ))
    ).resolve_axis(
        x='independent'
    ).resolve_scale(
        y='independent'
    ).configure_view(
        strokeWidth=0
    ).configure_facet(
        spacing=5
    ).configure_axis(
        grid=False
    )

    plot.show()
    return plot
```

```python
def make_ridgeline_plot(df, rawdf):
    """Create a ridgeline plot showing SGE score distributions by variant consequence."""
    
    # Define color palette and sort order
    PALETTE = {
        'Synonymous': '#006616',
        'Missense': '#81B4C7',
        'Stop Gained': '#ffcd3a',
        'Intron': '#6AA84F',
        'UTR Variant': '#93C47D',
        'Stop Lost': '#888888',
        'Start Lost': '#000000',
        'Canonical Splice': '#1170AA',
        'Splice Region': '#CFCFCF',
        '3bp Deletion': '#FF9A00'
    }
    
    VARIANT_ORDER = [
        'Synonymous', 'Missense', 'Stop Gained', 'Intron', 'UTR Variant',
        'Stop Lost', 'Start Lost', 'Canonical Splice', 'Splice Region', '3bp Deletion'
    ]
    
    # Set up the figure
    n_categories = len(VARIANT_ORDER)
    fig, axes = plt.subplots(n_categories, 1, figsize=(12, 8), sharex=True)
    
    # Plot each variant type
    for i, (ax, variant) in enumerate(zip(axes, VARIANT_ORDER)):
        # Get data for this variant type
        variant_data = df[df['Consequence'] == variant]
        
        if len(variant_data) > 0:
            # Plot the density curve
            ax.fill_between(
                variant_data['SGE_score'], 
                variant_data['density'],
                color=PALETTE[variant],
                alpha=0.7,
                edgecolor='black',
                linewidth=1.5
            )
        
        # Clean up axes
        ax.set_ylim(0, None)
        ax.set_ylabel(variant, rotation=0, ha='right', va='center', fontsize=14, labelpad=20)
        ax.set_yticks([])
        ax.spines['left'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        
        # Only show x-axis on bottom plot
        if i < n_categories - 1:
            ax.spines['bottom'].set_visible(True)
            ax.tick_params(axis='x', which='both', length=0)
        else:
            ax.spines['bottom'].set_visible(True)
            ax.set_xlabel('SGE Score', fontsize=16)
            ax.tick_params(axis='x', labelsize=14)
    
    # Adjust spacing
    plt.subplots_adjust(hspace=0.1)
    plt.show()
    return fig
```
